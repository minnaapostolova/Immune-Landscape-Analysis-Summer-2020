## Setting up Gene Lists for Cross Referencing

- genes list = list of all genes from data frame

```python
a = ["CD19", "MS4A1", "FOXP3", "IL2RA", "LRRC32", "TGFB1", "HLA-A", "HLA-B", "HLA-C", "HLA-DMA", "HLA-DMB", "HLA-DOA", "HLA-DOB", "HLA-DRB1", "ARG1", "ARG2", "CD33", "CSF2", "CSF3", "IDO1", "IL10", "IL13", "IL1B", "IL4", "IL6", "ITGAM", "CD274", "CD276", "CD28", "CD80", "CD86", "CTLA4", "HAVCR2", "LAG3", "LGALS9", "PDCD1", "PDCD1LG2", "VTCN1", "CD247", "CD3D", "CD3E", "CD3G", "CD4", "CD8A", "CD8B", "CD27", "CD70", "ICOS", "ICOSLG", "TNFRSF4", "TNFRSF9", "TNFSF9", "GZMA", "GZMB", "PRF1", "IFNG", "IL12A", "IL12B", "IL2", "LTA", "TNF", "KIR2DL1", "KIR2DL4", "KIR2DS4", "KIR3DL1", "KIR3DL2"]
len(a)
```

```python
genes = list(df['Hugo_Symbol'])
```

```python
indeces = []
for i in a:
    if i in genes:
        index = genes.index(i)
        indeces.append(index)
```

## Filtering Dataframe Based on "a" List

```python
df = df.loc[indeces]
print(df)
```
